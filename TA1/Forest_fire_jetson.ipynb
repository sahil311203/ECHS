```python
from ultralytics import YOLO
from google.colab import drive
import os

# --- MOUNT YOUR GOOGLE DRIVE ---
drive.mount('/content/drive', force_remount=True)

# --- SET YOUR SAVE LOCATION ---
project_save_path = "/content/drive/MyDrive/My_Fire_Training_Results"
experiment_name = "FireDetect-yolov8s-Run2" # Keeping the same name as your last attempt

# 1. Get the path to the data.yaml from the 'dataset' variable
# (Make sure you ran the cell to download your new dataset version first!)
data_yaml_path = os.path.join(dataset.location, "data.yaml")

print(f"--- Using dataset YAML from: {data_yaml_path} ---")
print(f"--- Saving results to: {project_save_path}/{experiment_name} ---")

# 2. Load the YOLOv8s (small) model.
model = YOLO('yolov8s.pt')

# 3. Start training!
results = model.train(
    data=data_yaml_path,
    model='yolov8s.pt',
    imgsz=640,
    epochs=50,  # <--- HERE IS THE CHANGE
    batch=16,
    project=project_save_path,
    name=experiment_name,
    exist_ok=True
)

print(f"--- Training Complete! (50 Epochs) ---")
print(f"Your trained model is saved in: {project_save_path}/{experiment_name}/weights/best.pt")
```
